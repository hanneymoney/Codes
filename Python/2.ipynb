# Explore average score by product

```python
avg_score_by_product = merged_df_all.groupby('Name')['Av_Score'].mean().sort_values(ascending=False)
avg_score_by_product.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Average Score by Product')
plt.xlabel('Product Name')
plt.ylabel('Average Score')
plt.show()
```

# Explore user distribution by state

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='State', data=users_df, palette='viridis')
plt.title('User Distribution by State')
plt.xlabel('State')
plt.ylabel('Number of Users')
plt.show()
```

# Explore year of birth distribution

```python
users_df['Year_of_Birth'] = pd.to_datetime(users_df['DOB'], format='%d.%m.%Y').dt.year
plt.figure(figsize=(12, 6))
sns.histplot(users_df['Year_of_Birth'], bins=30, kde=True, color='purple')
plt.title('Year of Birth Distribution')
plt.xlabel('Year of Birth')
plt.ylabel('Count')
plt.show()
```
